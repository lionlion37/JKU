```
import numpy as np
import pandas as pd
import altair as alt
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.base import clone
```

```
y_test = pd.read_pickle("data/y_test.pkl")
y_train = pd.read_pickle("data/y_train.pkl")

X_test = pd.read_pickle("data/X_test.pkl")
X_train = pd.read_pickle("data/X_train.pkl")
X_test_transformed = np.load("data/X_test_transformed.npy", allow_pickle=True)
X_train_transformed = np.load("data/X_train_transformed.npy", allow_pickle=True)
```

```
def graph_overfit(model_class,default_params,param, param_vals, scale="linear"):
    results = {
    "score":[],
    "Set Type":[],
    "param":[]
    }
    for param_val in param_vals:
        model = clone(model_class)
        model.set_params(**default_params)
        model.set_params(**{param:param_val})
        print(param, param_val)

        model.fit(X_train_transformed,y_train)
        train_score = model.score(X_train_transformed,y_train)
        results["score"].append(train_score)
        results["Set Type"].append("Train")
        results["param"].append(param_val)

        test_score = model.score(X_test_transformed,y_test)
        results["score"].append(test_score)
        results["Set Type"].append("Test")
        results["param"].append(param_val)
    return alt.Chart(pd.DataFrame(results)).mark_line(interpolate='basis').encode(
            x=alt.X("param",scale=alt.Scale(type=scale),title=param),
            y=alt.Y("score:Q",title="Accuracy",scale=alt.Scale(zero=False),axis=alt.Axis(format='%')),
            color="Set Type:N"
        )
```

```
C = [0.001,0.1,1,10,100,1000]
gamma = [0.0001,0.001,0.01,0.1,1,10]
degree = [2,3,4,5,6,7]

model=SVC()
default_params = {"kernel":"poly"}
(graph_overfit(model,default_params, "C",C,"log") | graph_overfit(model,default_params, "gamma",gamma,"log")) & \
(graph_overfit(model,default_params, "degree",degree))
```

```
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1000, num = 5)]

criterion = ["gini","entropy"]

max_depth = [int(x) for x in np.linspace(10, 110, num = 5)]
max_depth.append(None)

min_samples_split = [2, 5, 10]

min_samples_leaf = [1, 2, 4]

model = RandomForestClassifier()
default_params = {"criterion":"gini"}
(graph_overfit(model,default_params, "n_estimators",n_estimators) |graph_overfit(model,default_params, "max_depth",max_depth) )& \
(graph_overfit(model,default_params, "min_samples_split",min_samples_split)|graph_overfit(model,default_params, "min_samples_leaf",min_samples_leaf) )
```
