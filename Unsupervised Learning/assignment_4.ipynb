The cumulant-generating function $K_X(t)$ of a random variable $X$ is defined as 
the natural logarithm of the moment-generating function
\begin{equation}
  \newcommand\E{\operatorname{\mathbb{E}}}
  \newcommand\dif{\mathrm{d}}
  K_X(t) = \log \E\left[e^{tX}\right]. 
\end{equation}
Let $\kappa_n(X) = K_X^{(n)}(0)$ the $n$-th derivative of $K_X(t)$ evaluated at $t=0$, 
then the Maclaurin series is 
\begin{equation}
  K_X(t) = \sum_{n=1}^\infty \kappa_n(X) \frac{t^n}{n!}. 
\end{equation}

<h2 style="color:rgb(0,120,170)">Task 1 (20 Points)</h2>
Show that if the random variables $X$ and $Y$ are independent then 
$\kappa_n(X+Y) = \kappa_n(X) + \kappa_n(Y)$ and $\kappa_n(\alpha X) = \alpha^n \kappa_n(X)$
holds. 

$$\kappa_n(X+Y)=$$<br>
$$=\frac{d^n K_{X+Y}(t)}{d^nt}(0)=$$<br>
$$=\frac{d^n \log \mathbb{E}(e^{t(X+Y)})}{d^nt}(0)=^{\text{*(independence of X and Y)}}$$<br>
$$=\frac{d^n \log [\mathbb{E}(e^{tX})*\mathbb{E}(e^{tY})]}{d^nt}(0)=$$<br>
$$=\frac{d^n (\log \mathbb{E}(e^{tX}) + \log \mathbb{E}(e^{tY}))}{d^nt}(0)=$$<br>
$$=\frac{d^n\log \mathbb{E}(e^{tX}) + d^n\log \mathbb{E}(e^{tY})}{d^nt}(0)=$$<br>
$$=\frac{d^n\log \mathbb{E}(e^{tX})}{d^nt}(0)+\frac{d^n\log \mathbb{E}(e^{tY})}{d^nt}(0)=$$<br>
$$=\kappa_n(X) + \kappa_n(Y)$$<br>
<br><br>
$$\kappa_n(\alpha X)=K_{\alpha X}^{(n)}(0)=(K_{\alpha X}(t))^{(n)}(0) = (\log \mathbb{E}\left[e^{t\alpha X}\right])^{(n)}(0)=(K_{X}(\alpha t))^{(n)}(0)=$$<br>
$$=\left(\sum_{n=1}^\infty \kappa_n(X) \frac{(\alpha t)^n}{n!}\right)^{(n)}(0)=\left(\sum_{n=1}^\infty \kappa_n(X) \frac{\alpha^n t^n}{n!}\right)^{(n)}(0)=\alpha^n \left(\sum_{n=1}^\infty \kappa_n(X) \frac{t^n}{n!}\right)^{(n)}(0)=\alpha^n \kappa_n(X)$$

<h2 style="color:rgb(0,120,170)">Task 4 (20 Points)</h2>

Compare PCA and ICA at the example of the `cifar-10.npy` dataset. Load the dataset, perform PCA and ICA using the implementations of the `sklearn.decomposition` package with 100 components each. Then visualize the components extracted by the two methods using the provided functions `show_image` and `scale`. What differences do you observe and how do they relate to the properties of PCA and ICA? 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA, FastICA

def scale(s, a=0, b=1):
    return (b - a) * (s - s.min()) / (s.max() - s.min()) + a

def show_image(img):
    plt.imshow(img.reshape(3, 32, 32).transpose((1, 2, 0)))
```

```python
########## YOUR SOLUTION HERE ##########

data = np.load('cifar-10.npy')
ica = FastICA(n_components=100)
pca = PCA(n_components=100)

ica.fit(data)
pca.fit(data)
```

```python
fig, axs = plt.subplots(10,4)
fig.set_figwidth(10)
fig.set_figheight(30)

n = 0
for y in range(5):
    for x in range(4):
        axs[y,x].imshow(scale(pca.components_[n]).reshape(3, 32, 32).transpose((1, 2, 0)))
        axs[y,x].set_title(f'PCA component {n}')
        n += 1
        
n = 0
for y in range(5,10):
    for x in range(4):
        axs[y,x].imshow(scale(ica.components_[n]).reshape(3, 32, 32).transpose((1, 2, 0)))
        axs[y,x].set_title(f'ICA component {n}')
        n += 1
```

The PCs seem to be relatively evenly distributed and therefore have a high variance, which is the goal of PCA. But some components seem very similar (e.g. 5 and 7, 10 and 11) and therefore not independent of each other.

The ICs on the other hand don't seem to have this porperty (variance maximation is not objective of ICA). Each IC seems to highlight one specific, dominant, <b>independent</b> feature of the pictures.

Consider the cocktail party problem. Read the left channel of the files `birds.wav`, `ocean.wav`, `frogs.wav` using the `scipy.io.wavfile` package. Then construct a full-rank $3 \times 3$ mixing matrix and produce 3 different mixings of those signals. You can listen to them using the `sounddevice` package. Finally, use `FastICA` from the `sklearn.decomposition` package to recover the original sources. Listen to the recovered sources and compare them to the originals to check if ICA was successful. *Hint:* Scale the signals to $[-1,1]$ before playing them. 

```python
import scipy.io.wavfile
import sounddevice as sd

########## YOUR SOLUTION HERE ##########

birds = scipy.io.wavfile.read('birds.wav')[1][:,0]
ocean = scipy.io.wavfile.read('ocean.wav')[1][:,0]
frogs = scipy.io.wavfile.read('frogs.wav')[1][:,0]

source = np.c_[birds, ocean, frogs]
mixing_matrix_1 = np.array([[1, 1, 1], [0.5, 2, 1.0], [1.5, 1.0, 2.0]])
mixing_matrix_2 = np.array([[1, 0.8, 1], [1, 2, 0.01], [1, 2, 2]])
mixing_matrix_3 = np.array([[8, 8, 8], [0.02, 0.06, 0.9], [1, 0.2, 3]])

mix_1 = np.dot(source, mixing_matrix_1.T)
mix_2 = np.dot(source, mixing_matrix_2.T)
mix_3 = np.dot(source, mixing_matrix_3.T)

ica_1 = FastICA(n_components=3)
ica_2 = FastICA(n_components=3)
ica_3 = FastICA(n_components=3)

source_rec_1 = ica_1.fit_transform(mix_1)
source_rec_2 = ica_2.fit_transform(mix_2)
source_rec_3 = ica_3.fit_transform(mix_3)
```
